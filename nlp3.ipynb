```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification")
```

```python
res = classifier("I have constant pain in my joints.", candidate_labels = ["Cardiovascular", "Respiratory", "Neurologic", "Gastrointestinal", "Hepatobiliary", "Musculoskeletal", "Renal_Urologic", "Metabolic_Nutritional", "Immune Disorder", "Hematologic", "Endocrine", "Eye", "ENT", "Skin"],)
print(res)
```
